```python
from numpy import random
sem1_sub = np.random.randint(30,100, size=(30,4))
sem1_sub
```

```python
sem2_sub = np.random.randint(30,100, size=(30,4))
sem2_sub
```

```python
sem3_sub = np.random.randint(30,100, size=(30,4))
sem3_sub
```

```python
sem1 = pd.DataFrame(sem1_sub, columns = ["Physics","Chemistry","Math1","Civil"])
sem1
                                         
```

```python
passing_marks = 50
num_subjects = 4
total_marks = passing_marks * num_subjects
sem1["Total_score"] = sem1[["Physics","Chemistry","Math1","Civil"]].sum(axis=1)


# adding an extra columns to check pass/fail for each subject/
sem1["Physics_result"] = np.where(sem1['Physics'] >= passing_marks, "Pass", "Fail")
sem1["Chemistry_result"] = np.where(sem1['Chemistry'] >= passing_marks, "Pass", "Fail")
sem1["Math1_result"] = np.where(sem1['Math1'] >= passing_marks, "Pass", "Fail")
sem1["Civil_result"]= np.where(sem1["Civil"] >= passing_marks, "pass", "Fail")
```

```python
sem1["over_all"] = sem1.apply(lambda x: 'Pass' if (x['Total_score'] >= 200) else 'Fail', axis=1)
```

```python
sem1
```

```python
sem1["over_all"].value_counts()
```

```python
subject = []
p_count = []
f_count = []

for i in sem1.columns[-5:]:
    sem1_temp = sem1[i].value_counts()
    #append the list
    subject.append(i)
    p_count.append((sem1_temp.get(0) / sem1.shape[0])*100)
    f_count.append((sem1_temp.get(1) / sem1.shape[0])*100)
    
#plot

width = 0.35
fig, ax = plt.subplots(figsize=(12, 8))
p1 = ax.bar(subject, p_count, width, color='c')
p2 = ax.bar(subject, f_count, width, bottom=p_count, color='r')

# Naming the y-axis and the whole graph
plt.ylabel("Students (%)")
plt.title("Performance of Students")
  
# add tyticks and legend
ax.set_yticks(np.arange(0, 121, 15))
ax.legend(labels=['Pass', 'Fail'], loc='upper right')

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='center', fmt='%.1f')
ax.bar_label(p2, label_type='center', fmt='%.1f')

# graph_styling()

# To load the display window
plt.show()
```
